## Task 1

```python
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['target'] = iris.target

# Display the first few rows of the dataset
df_iris.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Summary statistics
print(df_iris.describe())

# Pairplot to visualize the relationships between features
sns.pairplot(df_iris, hue='target', markers=["o", "s", "D"])
plt.show()
```

```python
from sklearn.cluster import KMeans

# Fit K-means with different values of K
kmeans_results = {}
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_iris.drop('target', axis=1))
    kmeans_results[k] = kmeans.inertia_

# Elbow method plot
plt.plot(list(kmeans_results.keys()), list(kmeans_results.values()), 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
# Choose an optimal K (e.g., 3)
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df_iris['kmeans_labels'] = kmeans.fit_predict(df_iris.drop('target', axis=1))

# Visualize clusters
sns.pairplot(df_iris, hue='kmeans_labels', markers=["o", "s", "D"])
plt.show()
```

## Task 2

```python
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering

# Load the Wine dataset
wine = load_wine()
df_wine = pd.DataFrame(data=wine.data, columns=wine.feature_names)
df_wine['target'] = wine.target

# Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
wine_pca = pca.fit_transform(df_wine.drop('target', axis=1))

# Create a DataFrame with the PCA results
df_wine_pca = pd.DataFrame(data=wine_pca, columns=['PC1', 'PC2'])
df_wine_pca['target'] = df_wine['target']

# 2D scatter plot of the PCA results
sns.scatterplot(x='PC1', y='PC2', hue='target', data=df_wine_pca, palette='Set1')
plt.title('PCA of Wine Dataset')
plt.show()

# K-means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df_wine_pca['kmeans_labels'] = kmeans.fit_predict(df_wine_pca[['PC1', 'PC2']])

# Hierarchical Clustering
hc = AgglomerativeClustering(n_clusters=3)
df_wine_pca['hc_labels'] = hc.fit_predict(df_wine_pca[['PC1', 'PC2']])

# Visualize K-means Clustering results
sns.scatterplot(x='PC1', y='PC2', hue='kmeans_labels', data=df_wine_pca, palette='Set1')
plt.title('K-means Clustering on PCA Reduced Wine Dataset')
plt.show()

# Visualize Hierarchical Clustering results
sns.scatterplot(x='PC1', y='PC2', hue='hc_labels', data=df_wine_pca, palette='Set1')
plt.title('Hierarchical Clustering on PCA Reduced Wine Dataset')
plt.show()
```

## Task 3

```python
from sklearn.datasets import load_breast_cancer
from scipy.cluster.hierarchy import dendrogram, linkage

# Load the Breast Cancer dataset
cancer = load_breast_cancer()
df_cancer = pd.DataFrame(data=cancer.data, columns=cancer.feature_names)

# Apply hierarchical clustering
Z = linkage(df_cancer, 'ward')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title('Dendrogram for Breast Cancer Dataset')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

## Task 4

```python
from sklearn.datasets import load_digits
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Load the Digits dataset
digits = load_digits()
df_digits = pd.DataFrame(data=digits.data)

# Standardize the data
scaler = StandardScaler()
digits_scaled = scaler.fit_transform(df_digits)

# Apply DBSCAN
dbscan = DBSCAN(eps=1.0, min_samples=5)
df_digits['dbscan_labels'] = dbscan.fit_predict(digits_scaled)

# Visualize the results
plt.scatter(digits_scaled[:, 0], digits_scaled[:, 1], c=df_digits['dbscan_labels'], cmap='plasma')
plt.title('DBSCAN Clustering on Digits Dataset')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.show()
```
